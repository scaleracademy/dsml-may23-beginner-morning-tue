```python
df_aerofit = pd.read_csv('/content/aerofit.csv')
```

```python
from scipy.stats import f_oneway, ttest_ind

from statsmodels.graphics.gofplots import qqplot
```

```python
np.random.seed(40)

df_aerofit["random_group"] = np.random.choice(["g1", "g2", "g3"], size = len(df_aerofit))
```

```python
df_aerofit.head()
```

```python
sns.boxplot(x = "random_group", y = "Income", data = df_aerofit)
```

```python
income_g1 = df_aerofit[df_aerofit["random_group"] == "g1"]["Income"]
income_g2 = df_aerofit[df_aerofit["random_group"] == "g2"]["Income"]
income_g3 = df_aerofit[df_aerofit["random_group"] == "g3"]["Income"]
```

```python
income_g1.mean()
```

```python
income_g2.mean()
```

```python
income_g3.mean()
```

```python
# H0 -> All groups have same means
# Ha -> One or more groups have a different mean

f_stat, p_val = f_oneway(income_g1, income_g2, income_g3)
```

```python
f_stat
```

```python
p_val
```

```python
# Null Hypothesis Accepted
# F Stat low
# P Value higher than significance level
```

```python
# Shapiro Wilk Test

# if p-val < significance level -> Data not normally distributed
# p-val > significance level -> Data is normally distributed

# PROBLEM ->
# Fails if data is too large
```

```python
from scipy.stats import shapiro
```

```python
np.random.seed(40)
height_subset = height.sample(100)
```

```python
# H0: Data is gaussian
# Ha: Data is not gaussian
```

```python
test_stat, p_val = shapiro(height_subset)
```

```python
test_stat
```

```python
p_val
```

```python
np.random.seed(40)
df_wt_subset = df_wt.sample(100)["time"]
```

```python
test_stat, p_val = shapiro(df_wt_subset)
```

```python
test_stat
```

```python
p_val
```

```python
# LEVENE TEST
# Helps us determine whether the variances within the groups are roughly the same or not

# H0: Variances are equal
# Ha: Variances are not equal
```

```python
from scipy.stats import levene
```

```python
sns.histplot(data = df_hw, x = "Height", hue = "Gender")
```

```python
height_men = df_hw[df_hw["Gender"] == "Male"]["Height"]
height_women = df_hw[df_hw["Gender"] == "Female"]["Height"]
```

```python
levene_stat, p_val = levene(height_men, height_women)
```

```python
p_val
```

```python
# Since the p val are less than the significance level, we can reject our null hypothesis and say that the variances are not equal
```
